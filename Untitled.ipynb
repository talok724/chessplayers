```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("GM_players_statistics.csv")
```

```python
df["profile"] = df["profile"].apply(lambda x:x.split("/")[-1])
```

```python
df["name"] = df["name"].fillna(df["profile"])
```

```python
df.drop(columns=["Unnamed: 0","username","user_id","title","profile"],inplace=True)
```

```python
df.info()
```

```python
df
```

```python
df = df.dropna()
```

```python
date_columns = ["joined", "last_online", "highest_rapid_date", "highest_blitz_date", "highest_bullet_date", "highest_tactics_date"]

for col in date_columns:
    df[col] = pd.to_datetime(df[col])
    df[col] = df[col].dt.year
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import streamlit as st

# Assuming you have the DataFrame with the given data
data = pd.DataFrame({
    'current_rapid_rating': [2492.0],
    'highest_rapid_rating': [2566.0],
    'rapid_win': [179.0],
    'rapid_draw': [4.0],
    'rapid_loss': [9.0]
})

# Transpose the DataFrame to rotate the heatmap
data_transposed = data.T

# Create the heatmap using Seaborn's heatmap with the 'viridis' color palette
sns.heatmap(data_transposed, cmap='viridis', annot=True, fmt='.1f')

plt.xlabel('Rows')  # Swap the axis labels
plt.ylabel('Columns')

plt.title('Heatmap of DataFrame (Rotated 90 degrees)')

plt.tight_layout()  # To ensure the labels are not cut off

# Display the plot using st.pyplot
st.pyplot()
```
